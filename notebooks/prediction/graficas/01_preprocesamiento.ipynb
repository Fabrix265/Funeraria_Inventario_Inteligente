Celda 2: Carga de datos y diagnóstico inicial

```python
def cargar_datos(ruta):
    df = pd.read_excel(ruta)
    return df

def diagnostico_inicial(df):
    diag = {
        'shape': df.shape,
        'nulos_por_columna': df.isnull().sum().to_dict(),
        'dtypes': df.dtypes.astype(str).to_dict(),
        'n_registros': len(df),
        'n_columnas': len(df.columns),
    }
    return diag

df_raw = cargar_datos(PATH_INPUT)
diag = diagnostico_inicial(df_raw)

df = df_raw.drop(columns=[c for c in COLS_EXCLUIR if c in df_raw.columns])

if 'Monto' in df.columns:
    df['Monto'] = df['Monto'].replace('Vacío', np.nan)
    df['Monto'] = pd.to_numeric(df['Monto'], errors='coerce')

print(f"Dataset cargado: {diag['n_registros']} registros, {diag['n_columnas']} columnas originales.")
print(f"Dataset después de excluir PII: {df.shape[0]} registros, {df.shape[1]} columnas.")
print(f"Columnas actuales: {list(df.columns)}")
```

Celda 3: Tratamiento de Fechas

```python
def limpiar_fechas(df):
    log = {'imputadas': 0, 'sospechosas': 0, 'anomalias_anio': [], 'nros_imputados': []}

    if 'Nro' in df.columns:
        df = df.sort_values('Nro').reset_index(drop=True)

    df['Fecha'] = df['Fecha'].replace('Vacío', np.nan)
    df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y', errors='coerce')

    mask_nat = df['Fecha'].isna()
    fechas_nat = int(mask_nat.sum())
    if 'Nro' in df.columns:
        log['nros_imputados'] = df.loc[mask_nat, 'Nro'].tolist()

    df['Fecha'] = df['Fecha'].ffill()
    df['Fecha'] = df['Fecha'].bfill()
    log['imputadas'] = fechas_nat

    anio_modal = df['Fecha'].dt.year.mode()[0] if len(df) > 0 else 2025
    df['fecha_sospechosa'] = df['Fecha'].dt.year.apply(lambda x: abs(x - anio_modal) > 3)
    log['sospechosas'] = int(df['fecha_sospechosa'].sum())

    anomalias = df[df['Fecha'].dt.year != anio_modal]
    if len(anomalias) > 0:
        log['anomalias_anio'] = anomalias[['Fecha']].to_dict('records')

    df = df.sort_values('Fecha').reset_index(drop=True)

    return df, log

df, log_fechas = limpiar_fechas(df)

print(f"Fechas limpiadas: {log_fechas['imputadas']} imputadas, {log_fechas['sospechosas']} sospechosas.")
if log_fechas['nros_imputados']:
    print(f"  Nros con fecha imputada (revisar manualmente): {log_fechas['nros_imputados']}")
print(f"Rango de fechas: {df['Fecha'].min()} a {df['Fecha'].max()}")
```

Celda 4: Corrección de Monto y Outliers

```python
def limpiar_monto(df):
    log = {'outliers_detectados': 0, 'monto_original_207000': False}

    if 207000 in df['Monto'].values:
        df['Monto'] = df['Monto'].replace(207000, 2070.00)
        log['monto_original_207000'] = True

    Q1 = df['Monto'].quantile(0.25)
    Q3 = df['Monto'].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 3 * IQR

    df['monto_outlier'] = df['Monto'] > limite_superior
    log['outliers_detectados'] = int(df['monto_outlier'].sum())

    percentil_99 = df['Monto'].quantile(0.99)
    df['Monto_winsorizado'] = df['Monto'].clip(upper=percentil_99)

    return df, log

df, log_monto = limpiar_monto(df)

print(f"Monto corregido: outlier 207000 → 2070 = {log_monto['monto_original_207000']}")
print(f"Outliers detectados (IQR): {log_monto['outliers_detectados']}")
print(f"Percentil 99 para winsorización: {df['Monto_winsorizado'].max():.2f}")
```

Celda 5: Normalización de Categorías

```python
def normalizar_categoricas(df, mapas):
    log = {}
    for col, mapa in mapas.items():
        if col not in df.columns:
            continue
        antes = df[col].value_counts().to_dict()
        df[col] = df[col].astype(str).str.lower().str.strip()

        df_mapeado = df[col].map(mapa)
        cambios = int(df_mapeado.notna().sum())
        df[col] = df_mapeado.fillna(df[col])

        despues = df[col].value_counts().to_dict()
        log[col] = {'antes': antes, 'despues': despues, 'cambios': cambios}
    return df, log

mapas_normalizacion = {
    'Ataud_Modelo': ATAUD_MODELO_MAP,
    'Ataud_Color': ATAUD_COLOR_MAP,
    'Capilla': CAPILLA_MAP,
    'Forma de pago': PAGO_MAP,
}

df, log_cats = normalizar_categoricas(df, mapas_normalizacion)

for col, info in log_cats.items():
    print(f"{col}: {info['cambios']} valores normalizados")
```

Celda 6: Encoding de Binarias y Feature Engineering

```python
def encodear_binarias(df, cols):
    for col in cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip().str.lower().map({'si': 1, 'no': 0})
            df[col] = df[col].fillna(0).astype(int)
    return df

def verificar_cargadores(df):
    valores_unicos = df['Cargadores'].unique()
    print(f"Valores de Cargadores: {sorted(valores_unicos)}")
    return valores_unicos

def crear_ataud_completo(df):
    df['Ataud_completo'] = df['Ataud_Modelo'] + '_' + df['Ataud_Color']
    return df

df = encodear_binarias(df, COLS_BINARIAS)
verificar_cargadores(df)
df = crear_ataud_completo(df)

print("Binarias encodeadas (si=1, no=0) y feature 'Ataud_completo' creado.")
```

Celda 7: Agregación Mensual

```python
def agregar_mensual(df):
    df['Periodo'] = df['Fecha'].dt.to_period('M')

    df_mensual = df.groupby('Periodo').agg(
        servicios_totales=('Fecha', 'count'),
        monto_total=('Monto', 'sum'),
        monto_promedio=('Monto', 'mean'),
        monto_mediana=('Monto', 'median'),
        carroza_count=('Carroza', 'sum'),
        carroza_flores_count=('Carroza flores', 'sum'),
        auto_count=('Auto', 'sum'),
        microbus_count=('Microbus', 'sum'),
        cargadores_total=('Cargadores', 'sum')
    )

    for col_pivot in ['Ataud_Modelo', 'Ataud_Color', 'Capilla']:
        pivot = df.pivot_table(index='Periodo', columns=col_pivot, values='Monto', aggfunc='count', fill_value=0)
        pivot.columns = [f'{col_pivot}_{c}' for c in pivot.columns]
        df_mensual = df_mensual.join(pivot, how='left')

    df_mensual = df_mensual.fillna(0)

    all_months = pd.period_range(start=df['Periodo'].min(), end=df['Periodo'].max(), freq='M')
    df_mensual = df_mensual.reindex(all_months, fill_value=0)
    df_mensual.index.name = 'Periodo'
    df_mensual = df_mensual.reset_index()

    return df_mensual

df_mensual = agregar_mensual(df)

print(f"Dataset mensual: {len(df_mensual)} meses.")
print(f"Columnas: {list(df_mensual.columns[:10])}...")
```

Celda 8: Métricas de Calidad

```python
def calcular_metricas_calidad(df_original, df_limpio, df_mensual, logs):
    metricas = []

    n_modificados = logs.get('imputadas', 0)
    for col, info in logs.items():
        if isinstance(info, dict) and 'cambios' in info:
            n_modificados += info['cambios']
    n_registros = len(df_original)
    promedio_cambios = n_modificados / n_registros if n_registros > 0 else 0
    metricas.append(('Cambios totales aplicados', n_modificados))
    metricas.append(('Promedio de cambios por registro', f"{promedio_cambios:.2f}"))

    inconsistencias = 0
    for _, row in df_limpio.iterrows():
        if row['Ataud_Modelo'] == 'sin_ataud' and row['Cargadores'] > 0:
            inconsistencias += 1
    integridad = ((len(df_limpio) - inconsistencias) / len(df_limpio)) * 100 if len(df_limpio) > 0 else 0
    metricas.append(('Integridad de Datos', f"{integridad:.2f}%"))

    meses_rango = pd.period_range(start=df_limpio['Fecha'].min(), end=df_limpio['Fecha'].max(), freq='M')
    meses_presentes = set(df_mensual[df_mensual['servicios_totales'] > 0]['Periodo'])
    lagunas = len(meses_rango) - len(meses_presentes)
    metricas.append(('Consistencia Temporal', f"{lagunas} lagunas detectadas"))

    metricas.append(('Registros con Fecha Imputada', logs.get('imputadas', 0)))
    metricas.append(('Registros con Fecha Sospechosa', logs.get('sospechosas', 0)))

    n_outliers = df_limpio['monto_outlier'].sum()
    metricas.append(('Outliers en Monto', f"{n_outliers} ({(n_outliers/len(df_limpio))*100:.2f}%)"))

    for col, info in logs.items():
        if isinstance(info, dict) and 'cambios' in info:
            metricas.append((f'Categorías Normalizadas {col}', info['cambios']))

    fecha_min = df_limpio['Fecha'].min()
    fecha_max = df_limpio['Fecha'].max()
    n_meses = len(meses_rango)
    metricas.append(('Cobertura Temporal', f"{fecha_min.date()} a {fecha_max.date()} ({n_meses} meses)"))

    dist_mes = df_mensual[['Periodo', 'servicios_totales']]
    metricas.append(('Distribución por Mes', f"{len(df_mensual)} meses con datos"))

    return pd.DataFrame(metricas, columns=['Métrica', 'Valor'])

logs_combinados = {**log_fechas, **log_monto, **log_cats}
df_metricas = calcular_metricas_calidad(df_raw, df, df_mensual, logs_combinados)

print("Métricas de calidad calculadas:")
print(df_metricas.to_string(index=False))
```

Celda 9: Exportación de resultados

```python
def exportar_resultados(df_limpio, df_mensual, df_metricas, ruta_dir):
    df_limpio.to_excel(os.path.join(ruta_dir, 'dataset_limpio.xlsx'), index=False)
    df_mensual.to_excel(os.path.join(ruta_dir, 'dataset_mensual.xlsx'), index=False)
    df_metricas.to_excel(os.path.join(ruta_dir, 'metricas_preprocesamiento.xlsx'), index=False)
    print(f"Archivos exportados en: {ruta_dir}")

exportar_resultados(df, df_mensual, df_metricas, PATH_OUTPUT_DIR)
print("Pipeline de preprocesamiento completado.")
```

Celda 10: Pipeline Principal

```python
if __name__ == '__main__':
    df_raw = cargar_datos(PATH_INPUT)
    diag = diagnostico_inicial(df_raw)
    print(f"Datos cargados: {diag['n_registros']} registros, {diag['n_columnas']} columnas")

    df = df_raw.drop(columns=[c for c in COLS_EXCLUIR if c in df_raw.columns])
    if 'Monto' in df.columns:
        df['Monto'] = df['Monto'].replace('Vacío', np.nan)
        df['Monto'] = pd.to_numeric(df['Monto'], errors='coerce')
    df, log_fechas = limpiar_fechas(df)
    df, log_monto = limpiar_monto(df)
    df, log_cats = normalizar_categoricas(df, mapas_normalizacion)
    df = encodear_binarias(df, COLS_BINARIAS)
    df = crear_ataud_completo(df)
    df_mensual = agregar_mensual(df)

    logs_combinados = {**log_fechas, **log_monto, **log_cats}
    df_metricas = calcular_metricas_calidad(df_raw, df, df_mensual, logs_combinados)

    exportar_resultados(df, df_mensual, df_metricas, PATH_OUTPUT_DIR)
    print("Pipeline completado exitosamente.")
```
